## Цель и задачи исследования

Цель исследования заключается в получении модели, которая по некоторым данным о состоянии здоровья пациента сможет предсказать, есть ли у данного человека риск инсульта. 
Эта тема показалась мне интересной, потому что инсульт это очень опасное явление, которое может привести к смерти.


## Описание набора данных

Датасет представлен в виде одного .csv файла. Как собирались данные, кем и с какой целью - неизвестно. Размер ~300КБ.\
Выборка включает в себя 5110 записей и 12 признаков.
Целевая переменная **stroke** - означает, перенес ли пациент инсульт.
Признак id не несет полезной информации и его можно игнорировать.
- Количественные переменные:
    - **age** - возраст пациента;
    - **avg_glucose_level** - средний уровень сахара в крови;
    - **bmi** - индекс массы тела;
- Категориальные переменные:
    - **gender** - пол;
    - **hypertension** - есть ли у человека гипертония (1 - да, 0 - нет);
    - **heart_disease** - есть ли у человека сердечные заболевания;
    - **ever_married** - был ли когда-либо женат;
    - **work_type** - тип работы ("children" - дети, "Govt_job" - гос. служащий, "Never_worked" - никогда не работал, "Private" - частный работник, "Self-employed" - самозанятый);
    - **residence_type** - место проживания ("Rural" - сельская местность или "Urban" - город);
    - **smoking_status** - отношение к курению ("formerly smoked" - ранее курил, "never smoked" - никогда не курил, "smokes" - курит, "Unknown" - неизвестно);

## Генерация, нормализация признаков

Для лучшей работы модели все нечисловые признаки нужно по возможности привести к числовым.

```python
df["gender"] = df["gender"].map({"Male": 0, "Female": 1}); # 0 - мужчина, 1 - женщина
df["ever_married"] = df["ever_married"].map({"Yes": 1, "No": 0});
work_types = {"Govt_job": 0, "Never_worked": 1, "Private": 2, "Self-employed": 3, "children": 4}
df["work_type"] = df["work_type"].map(work_types);

df["Residence_type"] = df["Residence_type"].map({"Urban": 0, "Rural": 1});

smoking_statuses = {"formerly smoked": 0, "never smoked": 1, "smokes": 2, "Unknown": 3}
df["smoking_status"] = df["smoking_status"].map(smoking_statuses)

df.head(10)
```

```python
df.info()
```

Теперь  проведем нормализацию. Нормализация позволяет привести все числовые признаки в "равное положение". Все числовые признаки будут принимать значение от 0 до 1.

```python
def min_max_scaling(column):
    return (column - column.min())/(column.max()-column.min())

df["age"] = min_max_scaling(df["age"])
df["bmi"] = min_max_scaling(df["bmi"])
df["avg_glucose_level"] = min_max_scaling(df["avg_glucose_level"])
```

```python
df.head()
```

## Выбор метрики оценки качества

Посмотрим на распределение целевого признака.

В данной задаче очень важным показателем будет Recall у положительного класса. Эта метрика означает, какую долю из объектов класса мы угадали. Нам важно вычислить как можно больше людей из тех, у которых действительно есть предрасположенность к инсульту. На мой взгляд намного хуже, если человеку, которому правда грозит инсульт, сказать, что все в порядке, чем наоборот.

Наш датасет очень несбалансирован. В этом случае нельзя использовать метрику Accuracy, потому что даже самая глупая модель "У человека не будет инсульта" будет права в 95% случаев. В нашем случае корректно будет использовать метрику F1. Она объединяет в себе метрики Precision и Recall и хорошо подходит для несбалансированных выборок. Она принимает значения от 0 до 1 и близка к нулю, если Precision или Recall близки к нулю.

Будем обращать внимание на F1 и Recall положительного класса в частости.

## Разделение выборки

Разобьем выборку на тестовую и обучающую.

```python
X = df.drop("stroke", axis=1)
y = df["stroke"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

Так как наша выборка очень несбалансирована, применим SMOTE (Synthetic Minority Over-sampling Technique). SMOTE - это алгоритм предварительной обработки данных, используемый для устранения дисбаланса классов в наборе данных. В реальном мире нередко приходится обучать модель на наборе данных с очень малым количеством примеров определенного класса. Чаще всего эта проблема возникает при создании классификатора для диагностирования редких заболеваний, выявления производственных дефектов, раскрытия мошеннических транзакций. Одним из способов решения этой проблемы является сокращение числа примеров мажоритарного класса. Иными словами, из набора данных исключаются строки мажоритарного класса, чтобы выровнять количество строк мажоритарного и миноритарного классов.

Тем не менее при таком подходе теряется много данных, которые могли бы быть использованы для обучения модели и повышения ее.

Другим вариантом является увеличение примеров миноритарного класса. Иными словами, случайным образом дублируются образцы миноритарного класса. Проблема этого подхода заключается в том, что он приводит к переобучению, поскольку модель обучается на одних и тех же примерах.\
\
SMOTE помогает решить эту проблему. Он находит разность между данным образцом и его ближайшим соседом. Затем эта разность умножается на число в интервале от 0 до 1. Полученное значение добавляется к данному образцу для формирования нового синтезированного образца в пространстве признаков. Подобные действия продолжаются со следующим ближайшим соседом, до заданного пользователем количества образцов.

```python
from imblearn.over_sampling import SMOTE
smote = SMOTE()
X_train_resh, y_train_resh = smote.fit_resample(X_train, y_train.ravel())
```

### Случайный лес

Посмотрим на подробные данные о модели со случайным лесом.

```python
from sklearn.metrics import classification_report, plot_confusion_matrix
from sklearn.metrics import recall_score
from sklearn.metrics import accuracy_score
print(classification_report(y_test, rf_pred))
print('F1 Score: ',f1_score(y_test,rf_pred))
print('Recall Score: ',recall_score(y_test,rf_pred))
print('Accuracy Score: ',accuracy_score(y_test,rf_pred))
```

Видно, что имея большую accuracy, мы имеем плохой recall. Это означает, что модель очень редко угадывает людей, у которых на самом деле может быть инсульт.

Попробуем улучшить результат используя поиск по сетке.

```python
from sklearn.model_selection import GridSearchCV

n_estimators =[64,100,128,200]
max_features = [2,3,5,7]
bootstrap = [True,False]

param_grid = {'n_estimators':n_estimators,
             'max_features':max_features,
             'bootstrap':bootstrap}

rfc = RandomForestClassifier(random_state = 42)
grid = GridSearchCV(rfc,param_grid)
grid.fit(X_train,y_train)
grid.best_params_
```

Попробуем применить эти параметры.

```python
rfc = RandomForestClassifier(max_features=3,n_estimators=100,bootstrap=True, random_state=42)

rfc.fit(X_train_resh,y_train_resh)

rfc_tuned_pred = rfc.predict(X_test)

print(classification_report(y_test,rfc_tuned_pred))
print('F1 Score: ',f1_score(y_test,rfc_tuned_pred))
print("Recall score: ", recall_score(y_test, rfc_tuned_pred))
print('Accuracy Score: ',accuracy_score(y_test,rfc_tuned_pred))
```

Результат не изменился.

### Логистическая регрессия

Логистическая регрессия показывала лучший результат F1, попробуем с ней провести поиск по сетке.

```python
penalty = ['l1', 'l2']
C = [0.001, 0.01, 0.1, 1, 10, 100] 

log_param_grid = {'penalty': penalty, 
                  'C': C}
logreg = LogisticRegression(max_iter = 1000, solver='liblinear')
grid = GridSearchCV(logreg,log_param_grid)


grid.fit(X_train_resh,y_train_resh)


grid.best_params_
```

```python
logreg_clf = LogisticRegression(max_iter=1000, C=1, penalty='l2')
logreg_clf.fit(X_train_resh,y_train_resh)
logreg_tuned_pred = logreg_clf.predict(X_test)


print(classification_report(y_test,logreg_tuned_pred))

print('F1 Score: ',f1_score(y_test,logreg_tuned_pred))
print("Recall score: ", recall_score(y_test, logreg_tuned_pred))
print('Accuracy Score: ',accuracy_score(y_test,logreg_tuned_pred))
```

Мы видим, что логистическая регрессия после подбора гиперпараметров показывает лучший результат Recall по сравнению со случайным лесом, хоть и Accuracy стала чуть ниже. Тем не менее, F1 у нее выше, чем у случайного леса.

## XGBClassifier

 Попробуем применить классификатор XGBoost (Xtreme Gradient Boosting). Он имеет параметр, который помогает сбалансировать неравенство классов - scale_pos_weight.


Посмотрим распределение по целевому признаку в тренировочном датасете

```python
print(y_train.value_counts())
print()
print("scale_pos_weight:", 3416/160)
```

Построим модель

```python
from xgboost import XGBClassifier

xgb = XGBClassifier(random_state=42, scale_pos_weight=21.35)

xgb.fit(X_train, y_train)
xgb_pred = xgb.predict(X_test)

print(classification_report(y_test, xgb_pred))
print('F1 Score: ',f1_score(y_test,xgb_pred))
print('Recall Score: ',recall_score(y_test,xgb_pred))
print('Accuracy Score: ',accuracy_score(y_test,xgb_pred))
```

Мы видим результаты хуже, чем у логистической регрессии. Попробуем перебрать параметры по сетке.

```python
weights = [1,5,10,15,20, 21.35 ,25,30,35,40,50,100]
n_estimators =[64,100,128,200]
param_grid = dict(scale_pos_weight=weights, n_estimators=n_estimators)

xgb = XGBClassifier(random_state = 42)
grid = GridSearchCV(xgb,param_grid, scoring='f1')
grid.fit(X_train,y_train)
grid.best_params_
```

```python
xgb_tuned = XGBClassifier(random_state=42, scale_pos_weight=25, n_estimators=64)

xgb_tuned.fit(X_train, y_train)
xgb_tuned_pred = xgb_tuned.predict(X_test)

print(classification_report(y_test, xgb_tuned_pred))
print('F1 Score: ',f1_score(y_test,xgb_tuned_pred))
print('Recall Score: ',recall_score(y_test,xgb_tuned_pred))
print('Accuracy Score: ',accuracy_score(y_test,xgb_tuned_pred))
```

Полученный результат хуже, чем у логистической регрессии.
